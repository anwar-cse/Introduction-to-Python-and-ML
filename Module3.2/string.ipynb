String Tokenization

```python
s="i love country"
tokens=s.split()
print(tokens)
ss="what-is-phitron"
tok=ss.split("-")#  - er upor based kore split koreche
print(tok)

#joining
print(" ".join(tokens))
print("-".join(tokens))
```
