stack

```python
stack = []

stack.append(10)   # push
stack.append(20)
stack.append(30)

print(stack)   # [10, 20, 30]

stack.pop()    # remove last
print(stack)   # [10, 20]
```

queue

```python
from collections import deque

q = deque()

q.append(10)    # enqueue
q.append(20)
q.append(30)

print(q)   # deque([10, 20, 30])

q.popleft()   # dequeue
print(q)   # deque([20, 30])
```
