Indexing and slicing

```python
string="Anwar Hossain"
print(string[6])

s="python"
print(s[:])#full string
print(s[1:5:2])
print(s[:3]) #Start omit: means start index=0
print(s[2:]) #End omit
print(s[-1])#last index
```
