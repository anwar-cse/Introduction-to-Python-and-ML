```python
#arithmetic operation
a=10
b=3
print(a//b) #floor division
print(a**b) #power

#comparison operation
print(a<b) #False

#logical operation
print(a<20 and b>1) #true
print(not(a>9)) #false

#assignment operation
x=10
y=10
x**=2
print(x) #100
y//=3
print(y) #3

#bitwise operation
print(5 & 3, 5|3, 5^3, ~5) #AND=1, OR=7, XOR=6 ,NOT=-6
#bits left এ move (×2)
print(2<<3) #16
#bits right এ move (÷2)
print(10>>1)#5

#membership operation
z=[1,2,3]
print(2 in z) #True
print(5 not in z) #True

#identity operation
p=[1,2]
q=p
r=[1,2]
print(p is q) #true
print(p is r) #false
```
