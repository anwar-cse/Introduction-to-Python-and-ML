list and method

```python
#list
numbers=[100,10,20,3,4,5]
#access
print(numbers[2])

new_lst=numbers[0:3]
print(new_lst)
```

methods

```python
numbers = [10, 20, 30, 100]

# add
numbers.append(50)
print(numbers)

# insert
numbers.insert(0, 45)
print(numbers)

# delete last
numbers.pop()
print(numbers)

# delete specific
numbers.remove(100)
print(numbers)

# nested list
list_er_modde_list = [[1,3,2,4],[2,5,6,3]]
print(list_er_modde_list)


#sorted
numbers.sort()
print(numbers)
numbers.sort(reverse=True)
print(numbers)
```
