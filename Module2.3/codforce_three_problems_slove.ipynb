Min_Max:
https://codeforces.com/group/MWSDmqGsZm/contest/219158/problem/K

```python
inp=input()
num=inp.split()
a=int(num[0])
b=int(num[1])
c=int(num[2])

min=a
max=a

if(b<min):
    min=b
if(c<min):
    min=c
    
if(b>max):
    max=b
if(c>max):
    max=c
print(min, max)
```

check odd ,even, positive and negative
https://codeforces.com/group/MWSDmqGsZm/contest/219432/problem/C

```python
n=int(input())
inp=input().split()

odd=0; even=0; neg=0; pos=0;

for i in range(n):
    num=int(inp[i])
    if(num%2==0):
        even+=1
    else:
        odd+=1
    if(num<1):
        neg+=1
    else:
        pos+=1
    if(num==0):
        neg-=1
print("Even: ",even)
print("Odd: ",odd)
print("Positive: ",pos)
print("Negative: ",neg)
```

digits
https://codeforces.com/group/MWSDmqGsZm/contest/219432/problem/Q

```python
t=int(input())
for i in range(t):
    a=int(input())
    if(a==0):
        print(0)
        continue
    while a>0:
        k=a%10
        a//=10
        print(k,end=" ")
    print()
```
